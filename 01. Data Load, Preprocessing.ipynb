### 02.2 converted to wordnet’s format

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN



tokenized_fields['feel_overall_e_learning_tag'] = tokenized_fields['feel_overall_e_learning_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['access_device_tag'] = tokenized_fields['access_device_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['device_used_tag'] = tokenized_fields['device_used_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['area_data_availability_tag'] = tokenized_fields['area_data_availability_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['data_provider_tag'] = tokenized_fields['data_provider_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['time_tag'] = tokenized_fields['time_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['effective_of_remote_learning_tag'] = tokenized_fields['effective_of_remote_learning_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['teachers_helpful_tag'] = tokenized_fields['teachers_helpful_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['enjoy_remotely_tag'] = tokenized_fields['enjoy_remotely_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['stressful_distance_learning_tag'] = tokenized_fields['stressful_distance_learning_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['time_for_remotely_tag'] = tokenized_fields['time_for_remotely_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['contribution_of_internal_environment_tag'] = tokenized_fields['contribution_of_internal_environment_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['contribution_of_external_environment_tag'] = tokenized_fields['contribution_of_external_environment_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['satisfied_online_learning_tag'] = tokenized_fields['satisfied_online_learning_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
tokenized_fields['satisfied_manual_learning_tag'] = tokenized_fields['satisfied_manual_learning_tag'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])

tokenized_fields
```
